## Data analysis
In the following, relationships between attributes are examined.
### Videos per category
In the following graph, the distribution of trendy videos across the video categories was analyzed. Category 24 "Entertainment" was particularly impressive with a large number of assigned videos.

```python
sns.countplot(data=df, x='category_id')
```

### Likes per category
This graph visualizes the absolute number of likes for videos in a category. Videos in category 10 "music" are particularly noticeable with a large number of likes per video.


```python
sns.boxplot(data=df, x='category_id', y='likes', showfliers=False)
```

### Likes per rating
Here, the relative proportion of likes in all ratings given was examined. For the majority of the videos, the proportion of likes per rating is 90 to 98 percent.

```python
df['ratings'] = df['likes'] + df['dislikes']
df['likes_per_rating'] = df['likes'] / df['ratings']
df['likes_per_rating'].describe()
```

### Share of likes per rating for each category
At this point, it was examined whether there are any notable differences in the like-dislike distribution between the video categories. While this is mostly not the case, category 25 "News & Politicts" stands out due to its below-average positive ratio. The deviations in category 29 "Nonprofits & Activism" and 43 "Shows" are viewed as less meaningful due to the rather low proportion of videos in these categories.

```python
sns.boxplot(data=df, x='category_id', y='likes_per_rating', showfliers=False)
```

### Ratings per view
In addition to the like-dislike ratio, the share of reviews submitted per video view is to be examined at this point. For most videos, this percentage is between one and six percent, with a maximum of 42.7 ratings per hundred video views for a video.

```python
df['ratings_per_view'] = df['ratings'] / df['views']
df['ratings_per_view'].describe()
```

### Share of ratings per video view for each category
As with the like-dislike ratio, the relative share of ratings per video view is now considered individually for each category. No notable outliers were found here. The deviations of the categories 30 "Movies", 43 "Shows" and 44 "Trailers" have to be considered again with the relatively low number of videos in these categories in mind.

```python
sns.boxplot(data=df, x='category_id', y='ratings_per_view', showfliers=False)
```

### Views per category
Finally, the absolute number of video views of the individual categories was compared. Here category 10 "Music" stands out with high numbers. This coincides with the previous knowledge. The category already showed a high number of likes per video without it having particularly stood out in terms of the ratio of ratings to views.

```python
sns.boxplot(data=df, x='category_id', y='views', showfliers=False)
```
